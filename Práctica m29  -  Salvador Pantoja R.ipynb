```python
import requests
from bs4 import BeautifulSoup

url = 'https://listado.mercadolibre.com.mx/garmin#D[A:garmin]'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'lxml')

for item in soup.select('.ui-search-layout__item'):
    try:
        print('-------------------------')
        title = item.select_one('.poly-component__title-wrapper a')
        if title:
            print(title.get_text(strip=True))  
    except Exception as e:
        print(f'Error: {e}')
```

```python
import requests

url = "https://openlibrary.org/search.json?q=the+lord+of+the+rings"

response = requests.get(url)

if response.status_code == 200:
    print("Solicitud exitosa!")
    datos = response.json()

    libros = datos.get('docs', [])

    print(f"Total de libros encontrados: {len(libros)}")
    print("Listado de libros (ordenados por título):")
    for libro in sorted(libros, key=lambda x: x.get('title', '')):  
        titulo = libro.get('title', 'Título no disponible')
        autor = libro.get('author_name', ['Autor no disponible'])[0]  
        año = libro.get('first_publish_year', 'Año no disponible')
        print(f"Título: {titulo}")
        print(f"Autor: {autor}")
        print(f"Año de publicación: {año}")
        print("-" * 50)
else:
    print(f"Error: {response.status_code}")
```
